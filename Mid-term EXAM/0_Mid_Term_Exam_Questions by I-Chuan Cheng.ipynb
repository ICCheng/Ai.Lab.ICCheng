## Question 07 (10pt): Password Validation Algorithm

A website requests an user to input his account password. Write a program to examize the validity of the password? 

- The valid password must consists of:
    - At least 1 letter in \[a-z\]
    - At least 1 number in \[0-9\]
    - At least a letter in \[A-Z\] (capital)
    - At least one special letter in \[$#@!\]
    - At least 8 letters in total
    - At most 20 letters in total
- The program can accept a list of passwords that are split by commas.
- Print all valid passwords in a line and split by commas
- Examples:
    - Input: ABd1234@1,a F1#,2w3E*,2We3345
    - Output: ABd1234@1

````python
# Your code here!

import re
input_password = str(input("Enter passwords : "))
password_list = input_password.split(",")
password_valid=[]
for password in password_list:
    if(len(password)>8 and len(password)<20):
        if(re.search("[a-z]",password)):
            if(re.search("[A-Z]",password)):
                if(re.search("[0-9]",password)):
                    if(re.search("[$#@!]",password)):
                        password_valid.append(password)
print(','.join(map(str,password_valid)))
````

3. Move n-1 disks from auxiliary to destination.

```{r, tidy=FALSE, eval=FALSE, highlight=FALSE }
FUNCTION MoveTower(disk, source, destination, auxiliary):
IF disk == 0, THEN:
    move disk from source to destination
ELSE:
    MoveTower(disk - 1, source, auxiliary, destination)   // Step 1 above
    move disk from source to destination              // Step 2 above
    MoveTower(disk - 1, auxiliary, destination, source)   // Step 3 above
END IF
```
